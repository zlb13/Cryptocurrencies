# Clustering Crypto

```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
file_path = Path("crypto_data.csv")
crypto_df = pd.read_csv(file_path)
crypto_df.head()
```

```python
crypto_df.columns
```

```python
for column in crypto_df.columns:
    print(f"Column{column} has {crypto_df[column].isnull().sum()} null values")
```

```python
# Keep all the cryptocurrencies that are being traded.
new_crypto_df = crypto_df.loc[crypto_df["IsTrading"] == True]
new_crypto_df.head()
```

```python
# Remove the "IsTrading" column. 
df = new_crypto_df.drop(columns=["IsTrading"])
```

```python
# Remove rows that have at least 1 null value.

df = crypto_df.dropna()
df
```

```python
# Keep the rows where coins are mined.
df2 = df.loc[df["TotalCoinsMined"] > 0.0]
df2
```

```python
coin_name = df2["CoinName"]
coin_name
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
coin_name_df = pd.DataFrame({"CoinName": coin_name})
coin_name_df.reset_index(inplace=False)
coin_name_df.head()
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
final_crypto_df = df2.drop(columns=["CoinName", "IsTrading"])
final_crypto_df = final_crypto_df.set_index("Unnamed: 0")
final_crypto_df
```

```python
# Use get_dummies() to create variables for text features.
X = pd.get_dummies(final_crypto_df, columns=["Algorithm", "ProofType"])
X
```

```python
# Standardize the data with StandardScaler().
data_scaler = StandardScaler()
X_data_scaled = data_scaler.fit_transform(X)
X_data_scaled
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_data_scaled)
X_pca
```

```python
# Create a DataFrame with the three principal components.
pcs_df = pd.DataFrame(
    data=X_pca, columns=["PC 1", "PC 2", "PC 3"])

pcs_df.head()

```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.
inertia = []
k = list(range(1, 11))
# Calculate the inertia for the range of K values
for i in k:
   km = KMeans(n_clusters=i, random_state=0)
   km.fit(pcs_df)
   inertia.append(km.inertia_)
```

```python
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

Running K-Means with `k=4`

```python
# Initializing model with K = 4 
model = KMeans(n_clusters=4, random_state=5)

# Fitting model
model.fit(pcs_df)

# Get the predictions
predictions = model.predict(pcs_df)
print(predictions)

```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.

clustered_df = pd.concat([final_crypto_df, pcs_df], axis=1)
clustered_df.head()
```

```python
#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
clustered_df["Class"] = model.labels_
```

```python
#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies.
clustered_df["CoinName"] = coin_name_df
```

```python
# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```

### Deliverable 4: Visualizing Cryptocurrencies Results

#### 3D-Scatter with Clusters

```python
# Creating a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(
    clustered_df,
    x="CoinName",
    y="TotalCoinsMined",
    z="TotalCoinSupply",
    color="Class",
    symbol="Class",
    width=800,
    height=800,
    hover_name="CoinName",
   
   
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

```python
# Create a table with tradable cryptocurrencies.
clustered_df.hvplot.table(columns=['CoinName', 'Algorithm', 'ProofType', 'TotalCoinSupply', 'TotalCoinsMined', 'Class'], sortable=True, selectable=True)
```
